### A)

```python
import scipy.stats as st
import matplotlib.pyplot as plt

a = 0.6
b = 0.3**2

def beta_E(a, b):
    return round(a / (a+b),4)

def beta_var(a,b):
    return round((a*b)/(((a+b)**2) * (a+b+1)),4)

#print(f'E(Y): \t{beta_E(a,b)} \nVar(Y): {beta_var(a,b)}')

# Mas de facto, os parametros são a e b.

# Representação gráfica

x = np.linspace(0,1,100)
y = st.beta.pdf(x, a, b)

fig, ax = plt.subplots(1,1)

ax.plot(x, y, 'r-')
plt.title(f'Beta (a={a}, b={b})')

plt.show()
```

### B)

```python
new_a = a + 0.65*1000
new_b = b + 1000 - 0.65*1000

new_x = np.linspace(0,1,100)
new_y = st.beta.pdf(new_x, new_a, new_b)

fig, ax = plt.subplots(1,1)

plt.plot(x, y, 'r-')
plt.plot(new_x, new_y)
plt.legend(['Prior','Posterior'])
plt.title(f'Beta (a={new_a}, b={new_b})')

plt.show()

print(f'Prior:\nE(Y): \t{beta_E(a,b)} \nVar(Y): {beta_var(a,b)}\nPosterior:\nE(Y): \t{beta_E(new_a,new_b)} \nVar(Y): {beta_var(new_a,new_b)}')
```

My comment is that that with the new data, the distribution updated. In this case, it converged to a median of 0.65, with accordance with the probability given in the exercise text.  
Here lies some of the power of Bayesian statistics and it's power to update as new data arrives.

### C)
Let us , instead of fixing it in 65%, generate batches of size n, z times, with the probability between q% and w%

```python
new_new_a = a
new_new_b = b
```

```python
import random
batch_size = 500
iterations = 4
possibilites = np.linspace(0.45,0.85,100)

for i in range(0,iterations):
    sucesses = int(batch_size*random.choice(possibilites))
    new_new_a += sucesses
    new_new_b += batch_size - sucesses
    
new_new_x = np.linspace(0,1,100)
new_new_y = st.beta.pdf(new_new_x, new_new_a, new_new_b)

fig, ax = plt.subplots(1,1)

plt.plot(x, y, 'r-')
plt.plot(new_x, new_y)
plt.plot(new_new_x, new_new_y)
plt.legend(['Prior','Posterior', 'Posterior Random'])
plt.title(f'Beta (a={new_new_a}, b={new_new_b})')

plt.show()

print(f'Prior:\nE(Y): \t{beta_E(a,b)} \nVar(Y): {beta_var(a,b)}\n \
        \nPosterior:\nE(Y): \t{beta_E(new_a,new_b)} \nVar(Y): {beta_var(new_a,new_b)} \n\
        \nPosterior Random:\nE(Y): \t{beta_E(new_new_a,new_new_b)} \nVar(Y): {beta_var(new_new_a,new_new_b)}')
```
